```python
import cml.data_v1 as cmldata
import os
from pyspark.sql import SparkSession
from pyspark.sql.functions import lit
import pandas as pd


# Sample in-code customization of spark configurations
#from pyspark import SparkContext
#SparkContext.setSystemProperty('spark.executor.cores', '1')
#SparkContext.setSystemProperty('spark.executor.memory', '2g')

CONNECTION_NAME = "pdnd-prod-dl-1"
conn = cmldata.get_connection(CONNECTION_NAME)
spark = conn.get_spark_session()
print("Connection created")

# Sample usage to run query through spark
# EXAMPLE_SQL_QUERY = "show databases"
# spark.sql(EXAMPLE_SQL_QUERY).show()

target_table_name = "pagopa_dev.pos_v2_transaction_q2"
final_output_file = "Final_Enriched_POS_Data.csv"
```

```python
print(f"Step 1: Reading data from Iceberg table: {target_table_name}...")
# Read the table using Spark
df_target = spark.table(target_table_name)

print(f"Total rows to export: {df_target.count()}")
```

```python
print("Step 2: Collecting data to the driver (Converting to Pandas)...")
# .toPandas() pulls all 107k rows from the cluster into your Notebook's local memory
pdf_target = df_target.toPandas()
```

```python
print("Step 3: Writing local CSV file...")
# Use Pandas to write the CSV safely to your current Jupyter folder
pdf_target.to_csv(final_output_file, sep=";", index=False)

print(f"SUCCESS! You can now download '{final_output_file}' from your Jupyter sidebar.")
```
